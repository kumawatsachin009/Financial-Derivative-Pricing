## Poisson Path

POISSON DISTRIBUTION = Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event.

POISSON PROCESS = A Poisson Process is a model for a series of discrete event where the average time between events is known, but the exact timing of events is random . The arrival of an event is independent of the event before (waiting time between events is memoryless)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def poisson_Path(T=1, N=1000, paths=1, lamba=10):
    for i in range(paths):
        delta = float(T)/N
        t = np.arange(0, T+delta, delta)
        increments = np.random.poisson(lam=lamba*delta, size=N)
        cumsum = np.cumsum(increments)
        y = np.insert(cumsum, 0, 0)
        plt.plot(t, y)
```

```python
poisson_Path()
```

```python
poisson_Path(paths=5)
```

```python
def gamma_Path(T=1, N=1000, paths=1, lamba=10):
    for i in range(paths):
        delta = float(T)/N
        t = np.arange(0, T+delta, delta)
        increments = np.random.gamma(shape = 1, scale = 5*delta, size=N)
        cumsum = np.cumsum(increments)
        y = np.insert(cumsum, 0, 0)
        plt.plot(t, y)
```

```python
gamma_Path(paths=10)
```

## Brownian Motion

```python
def bm_paths(T=1, N=1000, paths=1):
    for i in range(paths):
        delta = float(T)/N
        t = np.arange(0, T+delta, delta)
        increments = np.random.normal(0, np.sqrt(delta), size=N)
        cumsum = np.cumsum(increments)
        y = np.insert(cumsum, 0, 0)
        plt.plot(t, y)
```

```python
bm_paths(paths=1)
```

```python
bm_paths(paths=10)
```

## GBM Paths

```python
def gbm_paths(T=1, N=1000, paths=1, drift=5):
    for i in range(paths):
        delta = float(T)/N
        t = np.arange(0, T+delta, delta)
        increments = np.random.normal(0, np.sqrt(delta), size=N)
        cumsum = np.cumsum(increments)
        y = np.insert(cumsum, 0, 0)
        plt.plot(t, np.exp(y)+drift*t)
```

```python
gbm_paths(paths=10)
```
